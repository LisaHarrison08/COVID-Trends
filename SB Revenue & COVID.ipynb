```python
# Dependencies and Setup
import json
import os
import pandas as pd
import urllib.request
import requests
from sqlalchemy import create_engine
```

```python
# Create reference to CSV file
csv_path2 = "Resources/Womply Revenue - State - Daily.csv"
womply_state_df = pd.read_csv(csv_path2, low_memory=False)
womply_state_df.head()
```

```python
#Pull Data from April to June
covid_revenue = womply_state_df.loc[womply_state_df["month"].isin([4,5,6])] 
covid_revenue
```

```python
#Pull data from states with highest cases: Texas, illinois, California, Wisconsin and Florida
revenue_byState = covid_revenue.loc[covid_revenue["statefips"].isin([48, 17, 6, 55, 12])] 
revenue_byState
```

```python
#Replace months & Statefips by Months and states names
revenue_byState["month"].replace({4:"April", 5: "May", 6: "June"}, inplace=True)
revenue_byState["statefips"].replace({48:"Texas", 17: "Illinois", 6: "California", 55: "Wisconsin", 12:"Florida"}, inplace=True)
print(revenue_byState)
```

```python
revenue_byState.head()
```

```python
#Query data types: Note that ss60 has some objects
revenue_byState.dtypes
```

```python
#Replace "." by 0 - create a new dataframe
covid_SB_rev =revenue_byState.replace(to_replace =".",
                 value =0) 
covid_SB_rev.head()
```

```python
#query data types to confirm changes - still shows as object but we should not have "." anymore
covid_SB_rev.dtypes
```

```python
#change ss60 data type
covid_SB_rev.revenue_ss60=covid_SB_rev.revenue_ss60.astype('float64')
```

```python
#query one last time
covid_SB_rev.dtypes
```

```python
covid_SB_rev.head()
```

```python
#Save Json file
SB_json_data = covid_SB_rev.to_json(r'C:\Users\ideuk\Desktop\ETL project\SB_data.json')
```

```python
#Save Dataframe into csv file
covid_SB_rev.to_csv('SB_data.csv', header=True, index=False) 
```

# COVID CASES

```python
# Create reference to CSV file
csv_path = "Resources/COVID - State - Daily.csv"

Covid_bystate_df = pd.read_csv(csv_path, low_memory=False)
Covid_bystate_df.head()
```

```python
#Drop COVID Columns - state
covid_cases_df = Covid_bystate_df.drop(columns=['death_count', 'new_death_count', 'test_count', 'case_rate', 'death_rate', 'test_rate', 'new_positives_rate', 'new_test_count', 'new_case_rate', 'new_death_rate', 'new_test_rate'])
covid_cases_df.head()
```

```python
covid_data = covid_cases_df.loc[covid_cases_df["month"].isin([4,5,6])] 
```

```python
covid_byState = covid_data.loc[covid_data["statefips"].isin([48, 17, 6, 55, 12])] 
covid_byState
```

```python
#Replace months & Statefips by Months and states names
covid_byState["month"].replace({4:"April", 5: "May", 6: "June"}, inplace=True)
covid_byState["statefips"].replace({48:"Texas", 17: "Illinois", 6: "California", 55: "Wisconsin", 12:"Florida"}, inplace=True)
print(covid_byState)
```

```python
#Query data types: Note that new case count has some objects
covid_byState.dtypes
```

```python
#Replace "." by 0 - create a new dataframe
covid_byState.replace(to_replace =".",
                 value =0) 
```

```python
covid_byState.new_case_count=covid_byState.new_case_count.astype('int64')
```

```python
covid_byState.dtypes
```
